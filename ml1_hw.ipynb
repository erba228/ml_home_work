```python
import pandas as pd
```

```python
import numpy
```

```python
import matplotlib.pyplot as plt
```

```python
file_path="data.csv"
data=pd.read_csv(file_path)
```

```python
print("Первые несколько строк данных:")
print(data.head())
```

```python
print("\nПропущенные значения в каждом столбце:")
print(data.isnull().sum())
```

```python
# Коэффициенты для конвертации
inches_to_meters = 0.0254
lbs_to_kg = 0.453592

# Конвертация и замена значений
data['Height'] = data['Height'] * inches_to_meters
data['Weight'] = data['Weight'] * lbs_to_kg

# Вывод данных после изменений
print("Измененные данные:")
print(data.head())
```

```python
# Вычисление среднего роста и веса
average_height = data['Height'].mean()
average_weight = data['Weight'].mean()

# Нахождение максимального и минимального роста
max_height = data['Height'].max()
min_height = data['Height'].min()

# Вывод результатов
print("Средний рост:", average_height, "м")
print("Средний вес:", average_weight, "кг")
print("Максимальный рост:", max_height, "м")
print("Минимальный рост:", min_height, "м")
```

```python
# Вычисление BMI
data['BMI'] = data['Weight'] / (data['Height'] ** 2)

# Вывод первых нескольких строк с добавленным столбцом BMI
print("Первые несколько строк с добавленным столбцом BMI:")
print(data.head())
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt;
```

```python
# Гистограммы распределения роста и веса
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(data['Height'], bins=20, color='skyblue', edgecolor='black')
plt.title('Распределение роста')
plt.xlabel('Рост')
plt.ylabel('Частота')
plt.subplot(1, 2, 2)
plt.hist(data['Weight'], bins=20, color='salmon', edgecolor='black')
plt.title('Распределение веса')
plt.xlabel('Вес')
plt.ylabel('Частота')
plt.show()
```

```python
# Точечный график с использованием seaborn
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Height', y='Weight', data=data)
plt.title('Точечный график роста и веса')
plt.xlabel('Рост')
plt.ylabel('Вес')
plt.show()
```

```python
# Ящичная диаграмма для роста с разделением по полу
plt.figure(figsize=(8, 6))
sns.boxplot(x='Gender', y='Height', data=data)
plt.title('Ящичная диаграмма роста с разделением по полу')
plt.xlabel('Пол')
plt.ylabel('Рост')
plt.show()
```

```python
# Построение графика распределения роста мужчин
plt.figure(figsize=(8, 6))
sns.histplot(data[data['Gender'] == 'Male']['Height'], color='skyblue', kde=True)
plt.title('Распределение роста мужчин')
plt.xlabel('Рост')
plt.ylabel('Частота')
plt.show()
```

```python
# Построение графика распределения роста женщин
plt.figure(figsize=(8, 6))
sns.histplot(data[data['Gender'] == 'Female']['Height'], color='salmon', kde=True)
plt.title('Распределение роста женщин')
plt.xlabel('Рост')
plt.ylabel('Частота')
plt.show()
```

```python
# График распределения по полу
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', data=data)
plt.title('Распределение по полу')
plt.xlabel('Пол')
plt.ylabel('Частота')
plt.show()
```

```python
import numpy as np

# Selecting only numeric columns
numeric_data = data.select_dtypes(include=[np.number])

# Изучение матрицы корреляции с помощью тепловой карты
plt.figure(figsize=(10, 8))
correlation_matrix = numeric_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Тепловая карта матрицы корреляции')
plt.show()
```
